```python
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df['Purchase Amount (USD)'] = df.groupby('Category')['Purchase Amount (USD)'].transform(lambda x: x.fillna(x.median()))
```

```python
df['Size'] = df.groupby('Category')['Size'].transform(
    lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else 'Unknown')
)
```

```python
df['Previous Purchases'] = df.groupby('Category')['Previous Purchases'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = [
    '_'.join(''.join(col.split('_')).lower().split())
    for col in df.columns
]

print(df.columns)
```

```python
df.columns = [
    'customer_id', 'age', 'gender', 'item_purchased', 'category',
    'purchase_amount', 'location', 'size', 'color', 'season',
    'review_rating', 'subscription_status', 'shipping_type',
    'discount_applied', 'previous_purchases', 'payment_method',
    'frequency_of_purchases'
]

print(df.columns)
```

```python
# create new column age_group
labels = ['young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly':90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)

print(df[['frequency_of_purchases', 'purchase_frequency_days']].head(10))
```

```python
df['promo_code_used'] = df['discount_applied']

df[['discount_applied', 'promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['discount_applied']).all()
```

```python
df = df.drop('promo_code_used', axis=1)
df.columns
```
